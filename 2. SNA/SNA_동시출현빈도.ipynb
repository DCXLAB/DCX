```python
okt = Okt() 

def preprocess_okt(text):
    #text = spacing(text) #띄어쓰기 보정 위에서 했으면 필요없습니다
    pos_words = okt.pos(text, stem=True)
    words = [word for word, tag in pos_words if tag in ['Noun', 'Adjective', 'Verb'] ]
    stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','아니다','이렇다','저렇다','도','다','돼다','최고','되다','를','같다','되었다','오다','가다','으로','자','에','와','한','하다','것','수','있다','않다','그','이다','이었다', '안','억','이건','정도','개']
    stopped_words = [w for w in words if not w in stopwords]
    return ', '.join(stopped_words)
```

```python
from tqdm import tqdm 
tqdm.pandas()
df1['review_token'] = df1['review'].progress_apply(preprocess_okt)
df2['review_token'] = df2['review'].progress_apply(preprocess_okt)
df3['review_token'] = df3['review'].progress_apply(preprocess_okt)
df4['review_token'] = df4['review'].progress_apply(preprocess_okt)
df5['review_token'] = df5['review'].progress_apply(preprocess_okt)
df6['review_token'] = df6['review'].progress_apply(preprocess_okt)
```

## 동시출현매트릭스 만들기 > 여기부턴 왔다갔다 하기...파일 섞이면 대참사

```python
df = pd.read_csv('total_token_임시저장.csv', encoding = 'utf-8-sig', index_col = 0)

import ast
from tqdm import tqdm

tqdm.pandas()

def literal_eval(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return val  # 변환에 실패한 경우 원래 값을 반환

df['token1'] = df['token1'].progress_apply(literal_eval)
total_token1 = df['token1'].dropna().tolist()
```

```python
# 동시출현빈도 매트릭스
cv=CountVectorizer()

X = cv.fit_transform(total_token1)
Xc = (X.T * X)
Xc.setdiag(0) #대각선 원소를 0으로
Xc=Xc.todense() # 밀집행렬로 변형 (matrix형으로)

names = cv.get_feature_names_out()

#Xc.toarray() .todense와 같지만 array형태로 만들어줘서 dataframe에 넣음
matrix = pd.DataFrame(data =Xc, columns = names, index = names)
matrix.index
```

## 만약 평균 기준으로 cutoff 한다면!
- 노드 빈도가 평균 이상일 경우만 살려주고 싶다면 사용하세요
- 데이터를 가볍게 만들길 원한다면,,

[reference]

하지만 빈도수 매트릭스의 경우 모든 키워드에 연결관계가 나타나 복잡할 가능성이 존재하기 때문에 매트릭스의 평균값을 기준(cut-off value)으로 평균보다 클 경우 1을, 작을 경우 0을 반환하여(slicing)새로운 바이너리 매트릭스(binary matrix)를 생성하였다(정풀잎 외, 2019; John Scott, 2012).

```python
mean_val = np.mean(Xc[Xc!=0]) # Xc의 평균값 계산
Xc[Xc < int(mean_val)] = 0 # 평균 이하인 값을 0으로 바꾸기
```

```python
# Xc를 numpy array로 변환
Xc = np.array(Xc)

# 각 행과 열의 합 구하기
row_sum = np.sum(Xc, axis=1)
col_sum = np.sum(Xc, axis=0)

# 합이 0인 인덱스 찾기
zero_row_idx = np.where(row_sum == 0)[0]
zero_col_idx = np.where(col_sum == 0)[0]

# 해당 인덱스 삭제
Xc = np.delete(Xc, zero_row_idx, axis=0)
Xc = np.delete(Xc, zero_col_idx, axis=1)

# 결과 출력
print(Xc)
Xc.shape
```

```python
names = cv.get_feature_names_out()
Name = [val for i, val in enumerate(names) if i not in zero_row_idx]
matrix = pd.DataFrame(data =Xc, columns = Name, index = Name)
matrix.index
```

```python
matrix.to_csv('co-occurrence matrix_cluster0.csv', encoding = 'utf-8-sig', sep = ',')
```

### 용량 너무 큰 애들 줄이기....

```python
matrix = pd.read_csv('co_occurrence matrix_cluster4.csv', encoding = 'utf-8-sig', index_col = 0)
```

```python
from ast import literal_eval

# 데이터프레임을 생성하는 예제 (이미 불러온 데이터프레임을 사용하세요)
# df = pd.read_csv('your_file.csv')

# 모든 셀에 대해 literal_eval을 적용하는 함수
def convert_to_numeric(cell):
    try:
        return literal_eval(cell)
    except (ValueError, SyntaxError):
        return cell

# 데이터프레임에 applymap 함수를 사용하여 각 셀을 변환
matrix = matrix.applymap(convert_to_numeric)

# 변환된 데이터프레임 인덱스만 출력
matrix.index
```

```python
# 각 키워드 동시출현 빈도의 합 계산
keyword_sums = matrix.sum(axis=1)

# 상위 n개의 키워드 > 승민 임의로 만개로 설정
n = 10000
top_keywords = keyword_sums.nlargest(n).index

# 상위 키워드에 해당하는 행과 열만 선택
reduced_co_occurrence_matrix = matrix.loc[top_keywords, top_keywords]

reduced_co_occurrence_matrix.index
```
