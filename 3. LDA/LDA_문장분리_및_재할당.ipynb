### LDA 키워드 선정하기
#### #장표에 있는 action 순서로 키워드 dictionary 만들어주기
*why? 추후 opportunity map 코드 돌릴때 덜 복잡함*
- LDA 키워드 선정 (확실한 키워드로) Action에 할당될 만한 키워드 추가
- topic에 할당될 만한 키워드 추가 (리뷰를 직접보고 확실한 키워드 추가)
- 아래 Action 순서 >> 장표 순서대로 나열합니다. 키워드는 각 Action에 사용한 topic의 주요 키워드를 적어봅니다.
- html(LDA 시각화 파일) 열어서 각 Action에 사용한 topic 누르고, 나오는 키워드들 중 상단부터 보면서 넣어봤습니다...

#### 헷갈릴 수 있으니 본인 담당 액션정리
- [topic2] / action1 식비지출
- [topic1,6] / action2 음식/반찬 고민. 정기배송
- [topic4] / action3 숙취, 설사, 변비
- [topic3,7] / action4 다이어트+과일
- [topic5,8] / action5 닭가슴살, 프로틴, 제로

```python
import pandas as pd
sent = pd.read_excel('식사챙김_전체문장분리_fn.xlsx', index_col = 0)
sent
```

```python
#LDA 키워드 #장표순서
action1 = ['비싸다','만원','돈','가격','싸다','물가','값','들다','외식','식당','쓰다','식비','아끼다']
action2 = ['맛있다','맛','야채','반찬','만들다','샐러드','도시락','메뉴','주문','배달','시키다','배송','정기','추천','귀찮다']
action3 = ['냄새','냉장고','유통','아프다','안좋다','숙취','죽다','괜찮다','변비','설사','화장실','위장','약국','수분','깨다','보관']
action4 = ['점심','다이어트','식단','단백질','계란','칼로리','고구마','살찌다','과일','딸기','바나나','토마토','스테비','요거트','방토','아침','저녁','탄수']
action5 = ['운동','헬스','산소','웨이','헬스장','유산','홈트','러닝','제로','프로','콜라','음료','우유','펩시','성분','낮다','당','대용','맛','칼로리','공복']

# 사전 만들기
name = [action1, action2, action3, action4, action5]

col_num = len(name)+1

dic={}
for n in range(1,col_num):
    dic[f'action_{n}'] = name[n-1]
dic
```

# 키워드 카운트

```python
sent_list = sent.review.values.ravel().tolist()

def calculate_keyword_frequency(data, keywords, topic_num):
    list_ = []
    for row in data:

        # total 초기화
        total = 0

        # 키워드별 빈도수를 저장할 딕셔너리 초기화
        frequencies = {keyword: 0 for keyword in keywords}

        for keyword in keywords:
            if keyword in row:
                frequencies[keyword] += 1

        # value값 다 더하기
        for value in frequencies.values():
            total += value

        list_.append(total)

    return pd.DataFrame({topic_num: list_})
```

```python
to = []
for i in range(1, col_num):
    result = calculate_keyword_frequency(sent_list, dic[f'action_{i}'], f'action_{i}')
    to.append(pd.DataFrame(result))
to.append(sent)

total = pd.concat(to, axis=1)



#-------------------------자신의 action 번호에 맞게 변경 필요---------------------------

# dictionary점수 모든 열이 0인 값 vs. 0이 아닌값을 분리
zero_rows = total[(total['action_1'] == 0)&(total['action_2'] == 0) &
                  (total['action_3'] == 0)&(total['action_4'] == 0) & (total['action_5'] == 0)]

non_zero_rows = total[~((total['action_1'] == 0)&(total['action_2'] == 0) &
                        (total['action_3'] == 0)& (total['action_4'] == 0) & (total['action_5'] == 0))]


#-------------------------------------------------------------



# 키워드 빈도 합이 가장 큰 값을 가지는 action을 최종 action으로 할당
col=[f'action_{i}' for i in range(1, col_num)]
max_column = non_zero_rows[col].idxmax(axis=1)

non_zero_rows['final_action_num']=[_[-1:] for _ in max_column]

print(non_zero_rows.final_action_num.value_counts())

display(non_zero_rows)
display(zero_rows)

print('재할당 후 reviewer 수: ',len(non_zero_rows.reviewer_id.unique()), '\n원래 전체 reviewer 수: ', len(sent))
```
