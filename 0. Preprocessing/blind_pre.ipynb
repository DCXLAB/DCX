### 정규식 -1

```python
import re

def extract_word(text) :
    hangul = re.compile('[^가-힣a-zA-Z]')
    result = hangul.sub(' ', text)
    return result
```

```python
clean_sheet['review'] = clean_sheet['review'].apply(lambda x : extract_word(x))
```

### 정규식-2

```python
import re

a = re.compile("\S*@\S+")
b = re.compile("#\S+")
c = re.compile("https?://\S+")
d = re.compile("www\.\S+")

def cleanse(text):
    result = a.sub("", text)
    result = b.sub("", result)
    result = c.sub("", result)
    result = d.sub("", result)
    result = result.replace("\u200b", "")
    result = result.replace("\n", "")
    return result
```

```python
clean_sheet['review'] = clean_sheet['review'].apply(lambda x : cleanse(x))
clean_sheet.reset_index(drop=True, inplace = True)
clean_sheet
```

### 소문자 변환 전처리

```python
corpus = []
for i in range(len(clean_sheet['review'])):
    review = clean_sheet['review'][i].lower() #lower case
    corpus.append(review)
clean_sheet['review'] = corpus
```

### 띄어쓰기 처리

```python
def re_spacing(text):
    result = text.strip()
    while "  " in result:
        result = result.replace("  ", " ")
    return result
```

```python
clean_sheet['review'] = clean_sheet['review'].apply(lambda x : re_spacing(x))
```
