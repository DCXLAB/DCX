```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
import numpy as np
from konlpy.tag import Okt
okt = Okt()
```

# 3D-CX 대표문장 불러오기

```python
DCX_=pd.read_excel('경험별 대표문장_total.xlsx')

#토크나이즈
computed_tok= [okt.morphs(i, stem = True) for i in DCX_['computed']]
connected_tok = [okt.morphs(i, stem = True) for i in DCX_['connected']]
shared_tok = [okt.morphs(i, stem = True) for i in DCX_['shared']]
shared_tok
```

# 자기 담당 원문 불러오기

```python
df1 = pd.read_csv("/Users/seungminkoo/Desktop/new_대학원/프로젝트/hy/3. 감성분석기회영역/[최종]new_touchpoint1_actor2_식사챙김_문장분리재할당_원문.csv", encoding = "utf-8-sig", index_col = 0)
df1.dropna(inplace = True)
df1.head()
```

# doc2vec 적용
- 담당원문

### 토큰화

```python
#stopwords
stopwords = pd.read_excel('ko-stopwords.xlsx')
stopwords=list(stopwords['stopwords'])
stopwords.extend(['에서','고','이다','는','한','씨', "것","거","게","데","이다","건","고","되다","되어다","걸","기",
                  "시","네","듯","랍니","중이","얘","스","도도", "나","수","개","내","기","제","저","인","있다","이렇다",
                  "그렇다","한다","하는","에","을","를","이","가","그래도","인","데","에서","이라서","요","고","의","과","가","까지",
                 "다","도","이라고","하다","하지만","그러나","그리고","그래서","그냥","합니다","도","는","을","넵","네","드랴","용",
                 "된","거","오","리","그러면","엔","저","도","세용","의","서","드","은","있네용","치고","슴","게","냐","뭉","된다는","중",
                 "성","넘나","용띠","아님", '넘','이라','인데','대해','해','있는데','하더라구','하더라고','하시면'])
stopwords=set(stopwords)


# 토큰화
okt=Okt()

def okt_pos_tagging(string):
    pos_words = okt.morphs(string)
    words = [word for word in pos_words if word not in stopwords ]
    return words

documents=[]
for i in tqdm(df1['sentence']):
    documents.append(okt_pos_tagging(i))

print(documents[:10])

#토큰 저장
df1['tokenizing'] = documents
df1.to_excel('[최종]식사챙김_token.xlsx')
df1.head()
```

### doc2vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import ast

tagged_doc = [TaggedDocument(words=doc, tags=[str(i)]) for i, doc in tqdm(enumerate(documents))]
```

```python
#doc2vec 모델 선언
model = Doc2Vec(vector_size=200, alpha=0.025, min_alpha=0.025, window=5)
model.build_vocab(x for x in tqdm(tagged_doc))
```

```python
#학습
for epoch in tqdm(range(30)):
    model.train(tagged_doc, total_examples=model.corpus_count, epochs=1)
    model.alpha -= 0.002
    model.min_alpha = model.alpha

# 모델 저장
model.save('doc2vec_model.doc2vec')
```

# 모델 infer

```python
#원문에 적용
df1["vector"] = df1["tokenizing"].progress_apply(lambda x : model.infer_vector(x))

#3D-CX에 적용
computed_vec = sum([model.infer_vector(i) for i in computed_tok])/len(computed_tok)
connected_vec = sum([model.infer_vector(i) for i in connected_tok])/len(connected_tok)
shared_vec = sum([model.infer_vector(i) for i in shared_tok])/len(shared_tok)
```

# 토픽 문장별 벡터 평균내기
- 변수조정필요
    - 자신의 action에 맞도록

```python
#토픽 unique확인
df1["topic"].unique()
```

```python
# 토픽 평균 내기
one_zero = sum(df1[df1["topic"] == 1]["vector"])/len(df1[df1["topic"] == 1]["vector"].index)
one_one = sum(df1[df1["topic"] == 2]["vector"])/len(df1[df1["topic"] == 2]["vector"].index)
one_two = sum(df1[df1["topic"] == 3]["vector"])/len(df1[df1["topic"] == 3]["vector"].index)
one_three = sum(df1[df1["topic"] == 4]["vector"])/len(df1[df1["topic"] == 4]["vector"].index)
one_four = sum(df1[df1["topic"] == 5]["vector"])/len(df1[df1["topic"] == 5]["vector"].index)
```

# 코사인 유사도 적용
- 자신의 액션에 맞도록 수정

```python
from scipy import spatial

# 코사인 유사도 구하는 함수 정의
def cosine_similarity(v1, v2):
    return abs(1 - spatial.distance.cosine(v1, v2))

rep = [computed_vec, connected_vec, shared_vec]

#자신의 액션에 맞도록 수정 > 액션 개수 맞춰서!!!!!!
real = [one_zero, one_one, one_two, one_three, one_four] # action 5개
# real = [one_zero, one_one, one_two, one_three] # action 4개
# real = [one_zero, one_one, one_two] # action 3개
# real = [one_zero, one_one] # action 2개

res = []
for i in rep:
    for w in real:
        res.append(cosine_similarity(i, w))
res
```

# 매트릭스로 예쁘게 만들기
- 변수조정필요
    - 자신의 액션에 맞도록

```python
#Touchpoint-Actor-Action 순으로 T-A-A 뒤에 숫자 넣기

idx = ["T1-A2-A1", "T1-A2-A2", "T1-A2-A3", "T1-A2-A4", "T1-A2-A5"]

# action의 개수만큼 간격 나누기 (ex. action이 4까지 있다면..간격은 4가 됨)
result = pd.DataFrame({"computed" : res[0:5], "connected" : res[5:10], "shared" : res[10:15]})
result.index = idx
result
```
